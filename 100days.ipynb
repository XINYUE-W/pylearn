```python
num = int(input("Give me one number"))
end = int (num ** 0.5)
is_prime = True
for i in range (2, end + 1):
        if num % i == 0:
              is_prime = False
              break
if is_prime:
    print(f'{num} is prime number')
else:
    print(f'{num} is not prime number')
```
