```python
import math

weight_input = input("Weight: ")

if not weight_input.isdigit():
    print("NOT VALID")

else:
    weight = int(weight_input)
    unit = input("(L)bs or (K)g: ").upper()
    if unit == "L":
        converted = math.ceil(weight * 0.45)
        print(f"You are {converted} kilos")
    elif unit == "K":
        converted = math.ceil(weight / 0.45)
        print(f"You are {converted} pounds")
    else:
        print("Invalid unit, please enter 'L' or 'K'.")
```

```python
secret_number = 9
guess_count = 0
guess_limit = 3

while guess_count < guess_limit:
    guess = input("Guess: ")
    
    if not guess.isdigit():     # æ£€æŸ¥è¾“å…¥æ˜¯å¦ä¸ºæ•°å­—
        print("NOT VALID â€” please enter a number.")
        continue                # é‡æ–°è¿›å…¥å¾ªçŽ¯ï¼Œä¸å¢žåŠ æ¬¡æ•°
    guess = int(guess)
    guess_count += 1
    if guess == secret_number:
        print ("YOU WIN")
        break
    else: 
        print ("WRONG")
else:
    print ("NO CHANCE")
```

```python
command = ""
started = True
while command.lower()!= "quit":
    command = input (">")
    if command.lower() == "start":
        if started:
            print ("The car has already started!")
        else:
            started = True
            print ("The car is ready to start")
    elif command.lower() == "stop":
        if not started:
            print ("The car has already stopped!")
        else:
            started = False
            print ("Car stopped")
    elif command.lower() == "help":
        print ("""start - to start the car
stop - to stop the car
quit - to quit the page""")
    elif command.lower() == "quit": 
        break
    else:
        print ("Sorry, I dont understand.")
```
